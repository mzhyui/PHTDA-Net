```python
import torch
from transformers import DistilBertModel, DistilBertTokenizerFast

# Load the pre-trained BERT model and tokenizer
model = DistilBertModel.from_pretrained('./data/bert_model/distil', output_attentions=True, output_hidden_states=True)
tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')

# Input text
text = "Example input sentence"

# Tokenize the text
tokens = tokenizer.encode_plus(text, add_special_tokens=True, return_tensors='pt')

# Forward pass through the model
with torch.no_grad():
    outputs = model(**tokens)

print(outputs)
# Get the hidden states
hidden_states = outputs.hidden_states
print(hidden_states)mz

# Get the middle hidden state
middle_state = hidden_states[len(hidden_states) // 2]

print("Shape of the middle hidden state:", middle_state.shape)
```
